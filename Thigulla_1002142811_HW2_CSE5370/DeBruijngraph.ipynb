```python
import networkx as nx
import matplotlib.pyplot as plt

#Reading in the reads and  kmers from the generated files
f = open("1002142811.txt", "r")
reads = [line.strip() for line in f]
f.close()

f = open("kmers.txt", "r")
k_1mers = [line.strip() for line in f]
f.close()

k = 2

#Generating the De Bruijn graph
graph = nx.DiGraph()
for kmer in k_1mers:
    graph.add_node(kmer)

for read in reads:
    for i in range(len(read)-k+1):
        kmer1 = read[i:i+k]
        kmer2 = read[i+1:i+k+1]
        if kmer1 in graph.nodes() and kmer2 in graph.nodes():
            graph.add_edge(kmer1, kmer2)

pos = nx.circular_layout(graph)
nx.draw(graph, pos, with_labels=True, node_size=1100, node_color='lightblue', font_size=12, font_weight='bold', font_family='times new roman')
nx.draw_networkx_edges(graph, pos, edge_color='black')
plt.axis('off')
plt.show()
```
